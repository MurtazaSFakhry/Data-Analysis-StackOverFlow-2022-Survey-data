```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(17, 10))

# Assuming you have the 'new_data' DataFrame

# Filter out rows with NaN values in 'Age' and 'EdLevel' columns
filtered_data = new_data[['Age', 'EdLevel']].dropna()

# Extract the 'Age' and 'EdLevel' columns as separate arrays
age = filtered_data['Age']
ed_level = filtered_data['EdLevel']

# Add 'Not specified' category to 'ed_level' array
ed_level = ed_level.append(pd.Series(['Not specified']))

# Get unique age groups and education levels
age_groups = age.unique()
education_levels = ed_level.unique()

# Count the occurrences of each education level for each age group
count_matrix = np.zeros((len(age_groups), len(education_levels)), dtype=int)

for i, age_group in enumerate(age_groups):
    age_group_data = filtered_data[age == age_group]
    count = age_group_data['EdLevel'].value_counts()
    count_matrix[i] = [count.get(level, 0) for level in education_levels]

# Create the bar chart
fig, ax = plt.subplots()

bar_width = 0.35
index = np.arange(len(age_groups))

for i, level in enumerate(education_levels):
    ax.bar(index + i * bar_width, count_matrix[:, i], bar_width, label=level)

# Set labels and title
ax.set_xlabel('Age Groups')
ax.set_ylabel('Count')
ax.set_title('Education Level by Age')
ax.set_xticks(index)
ax.set_xticklabels(age_groups, rotation=45)
ax.legend(fontsize = 'small')

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20, 12))  # Increase the overall chart size

# Assuming you have the 'new_data' DataFrame

# Filter out rows with NaN values in 'Age' and 'EdLevel' columns
filtered_data = new_data[['Age', 'EdLevel']].dropna()

# Extract the 'Age' and 'EdLevel' columns as separate arrays
age = filtered_data['Age']
ed_level = filtered_data['EdLevel']

# Add 'Not specified' category to 'ed_level' array
ed_level = ed_level.append(pd.Series(['Not specified']))

# Get unique age groups and education levels
age_groups = age.unique()
education_levels = ed_level.unique()

# Count the occurrences of each education level for each age group
count_matrix = np.zeros((len(age_groups), len(education_levels)), dtype=int)

for i, age_group in enumerate(age_groups):
    age_group_data = filtered_data[age == age_group]
    count = age_group_data['EdLevel'].value_counts()
    count_matrix[i] = [count.get(level, 0) for level in education_levels]

# Create the bar chart
fig, ax = plt.subplots()

bar_width = 0.5  # Increase the width of the bars for better spacing
index = np.arange(len(age_groups))

for i, level in enumerate(education_levels):
    ax.bar(index + i * bar_width, count_matrix[:, i], bar_width, label=level)

# Set labels and title
ax.set_xlabel('Age Groups')
ax.set_ylabel('Count')
ax.set_title('Education Level by Age')
ax.set_xticks(index)
ax.set_xticklabels(age_groups, rotation=45)
ax.legend(fontsize='small')

# Adjust spacing between bars
ax.set_xlim(-bar_width, len(age_groups) - bar_width)

# Display the chart
plt.tight_layout()
plt.show()
```
