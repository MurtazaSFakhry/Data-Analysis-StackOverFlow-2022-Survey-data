```python
import pandas as pd
import matplotlib.pyplot as plt

# Split the 'LanguageHaveWorkedWith' column into separate languages
languages = data["LanguageHaveWorkedWith"].str.split(';', expand=True)

# Count the occurrences of each language
language_counts = languages.stack().value_counts()

# Create a scatter plot
plt.figure(figsize=(12, 6))
#plt.scatter(language_counts.index, language_counts.values, color='#1f77b4', alpha=0.7, s=100)
plt.scatter(language_counts.index, language_counts.values, c=range(len(language_counts)), cmap='magma', alpha=0.7, s=100)

# Set the labels and title
plt.xlabel('Language', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Language Have Worked With', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.3)

# Set the background color of the plot
plt.gca().set_facecolor('#f2f2f2')

# Add data labels for each point
# for language, count in zip(language_counts.index, language_counts.values):
#     plt.text(language, count, f"{count:,}", ha='center', va='bottom', fontsize=8, color='black')

# Adjust plot margins


# Remove the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Set the axis labels' color
plt.gca().xaxis.label.set_color('#555555')
plt.gca().yaxis.label.set_color('#555555')

# Set the tick labels' color
plt.gca().tick_params(axis='x', colors='#555555')
plt.gca().tick_params(axis='y', colors='#555555')

# Set the color of the grid lines
plt.grid(color='#FFFFFF', linestyle='--', linewidth=0.5)

# Set the font style of the title
plt.title('Languages Have Worked With', fontsize=16, fontweight='bold')

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Split the 'LanguageWantToWorkWith' column into separate languages
languages = data['LearnCode'].str.split(';', expand=True)
languages_online = data['LearnCodeOnline'].str.split(';', expand=True)

# Count the occurrences of each language
language_counts = languages.stack().value_counts()
language_counts_online = languages_online.stack().value_counts()

# Create a scatter plot
plt.figure(figsize=(15, 10))

# Scatter plot for LearnCode
plt.scatter(language_counts.index, language_counts.values, c='red', alpha=0.7, s=100, label='LearnCode')

# Scatter plot for LearnCodeOnline
plt.scatter(language_counts_online.index, language_counts_online.values, c='orange', alpha=0.7, s=100, label='LearnCodeOnline')

# Set the labels and title
plt.xlabel('Resources', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('LearnCode and LearnCodeOnline', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=10)

# Add grid lines
plt.grid(True, linestyle='--', alpha=0.3)

# Set the background color of the plot
plt.gca().set_facecolor('#f2f2f2')

# Remove the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Set the axis labels' color
plt.gca().xaxis.label.set_color('#555555')
plt.gca().yaxis.label.set_color('#555555')

# Set the tick labels' color
plt.gca().tick_params(axis='x', colors='#555555')
plt.gca().tick_params(axis='y', colors='#555555')

# Set the color of the grid lines
plt.grid(color='#FFFFFF', linestyle='--', linewidth=0.5)

# Set the font style of the title
plt.title('LearnCode and LearnCodeOnline', fontsize=16, fontweight='bold')

# Add legends
plt.legend()

# Tight layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()
```
